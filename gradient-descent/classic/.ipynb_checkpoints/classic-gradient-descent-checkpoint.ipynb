## Gradient descent

$$
y = wx
$$


$$
Loss = \frac{1}{N} \sum(y - pred)^2
$$

$$
w(k+1) = w(k) - \nabla w(k)
$$

$$
\nabla w = \mu * \frac{\partial Loss }{\partial w} 
$$


$$
\frac {\partial Loss}{\partial w} = \frac {\partial Loss}{\partial y} \frac {\partial y}{\partial w}  
$$

$$
\frac {\partial Loss}{\partial y} = \frac {\partial (y - pred)^2}{\partial y} = 2 (y - pred)
$$

$$
\frac {\partial y}{\partial w} = x
$$

$$
\frac {\partial Loss}{\partial w} = 2(y - pred) * x
$$

$$
\nabla w = \mu * \frac{1}{N}\sum{2(y - pred)x} = \mu * \frac{2}{N}\sum_{i}{(y_{i} - pred_{i}) * x_{i}} 
$$


```python
def gradient(x, w, t):
    return 2 * x * ( x*w -t )
```

```python
def delta(w, x, t, learning_rate=0.9):
    return np.mean(gradient(x, w, t)) * learning_rate
```

```python
def loss(y, pred):
    return np.mean((y - pred)**2)
```

```python
w = np.random.rand()
```

```python
nb_of_iterations = 6
```

```python
w_loss = [(w, loss(input_x_with_noise*w, f(input_x_with_noise)))]
w_loss
```

```python
for i in range(nb_of_iterations):
    dw = delta(w, input_x*w, input_x_with_noise)
    w = w - dw
    w_loss.append((w, loss(input_x*w, input_x_with_noise)))

w_loss
```
